```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
#number of r&c
df.shape
```

```python
#statistical measures
df.describe()
```

```python
df['Outcome'].value_counts()
```

```python
df.groupby('Outcome').mean()
```

```python
#data seperate
X = df.drop(columns = 'Outcome', axis = 1)
Y = df['Outcome']
```

```python
print(X)
print(Y)
```

Standardization

```python
#ranges are too complex
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_df = scaler.transform(X)
```

```python
print(standardized_df)
```

```python
X = standardized_df
Y = df['Outcome']
```

```python
print(X)
print(Y)
```

Train Test Split

stratify yapmazsak tüm diabetlilerin x traine vs gitme ihtimali olabilir !!!

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training
classifier.fit(X_train, Y_train)
```

Evaluation

```python
#accuracy on training set
X_trainaccuracy = classifier.predict(X_train)
training_accuracy = accuracy_score(X_trainaccuracy, Y_train)
print("Accuracy of training set: ", training_accuracy)
```

```python
#accuracy on test set
X_testaccuracy = classifier.predict(X_test)
training_accuracy2 = accuracy_score(X_testaccuracy, Y_test)
print("Accuracy of test set: ", training_accuracy2)
```

Predictive System

```python
input_data = (5,116,74,0,0,25.6,0.201,30)
#changing as array
input_data_array = np.asarray(input_data)
input_data_reshape = input_data_array.reshape(1,-1)

standardized_data = scaler.transform(input_data_reshape)
print(standardized_data)

predict = classifier.predict(standardized_data)
print(predict)

if (predict[0] == 0):
  print("Non Diabetic")
else:
  print("Diabetic")
```
