```python
#separate datas
X = df.drop('quality', axis = 1)
```

```python
print(X)
```

```python
# want to classify 7+ good 7- bad quality(label encoding/binarization)
Y = df['quality'].apply(lambda y_value: 1 if y_value >=7 else 0)
```

```python
print(Y)
```

```python
#train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y,random_state = 2)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
#evaluation
X_test_predict = model.predict(X_test)
x_test_accuracy = accuracy_score(X_test_predict, Y_test)
```

```python
print('Accuracy: ', x_test_accuracy)
```

```python
input = (8.3,0.655,0.12,2.3,0.083,15.0,113.0,0.9966,3.17,0.66,9.8)
#change data to array
input_array = np.asarray(input)
input_array_reshape = input_array.reshape(1,-1)
prediction = model.predict(input_array_reshape)
print(prediction)

if(prediction[0]==1):
  print("It is good..!")
else:
    print("It is bad..!")
```
